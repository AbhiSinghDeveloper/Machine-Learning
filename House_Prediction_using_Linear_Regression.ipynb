```
#Load the House sales Dataset
#SFrame is a data struture for representing data in TURICREATE


sales = turicreate.SFrame('/content/drive/My Drive/Colab Notebooks/home_data.sframe')
```

```
sales
```

```
#Exploring the data for Housing Sales
#show() is used to explore the data

sales.show()
```

```
turicreate.show(sales[1:5000]['sqft_living'],sales[1:5000]['price'])
```

```
#Create a Simple Regression model of sqft_living to price
#Define the data set into two i.e. one for training and the other one for testing as shown below
#random_split randomly splits the SFrame into two SFrames(it generates pseudo random generation process)


train_data,test_data = sales.random_split(.8,seed=0)
```

```
#Build the Regression model
#turicreate.linear_regression is a method for creating and using the Linear Regression models


sqft_model = turicreate.linear_regression.create(train_data, target = 'price', features=['sqft_living'])
```

```
#Evaluate a simple model

#Calculating the mean value of the "price"

print(test_data['price'].mean())


 
```

```
#evaluate() is used to take a dataset and print out or retrun some statistics of how well that fit is doing
#here RMSE is Root Mean Square Error
print(sqft_model.evaluate(test_data))
```

```
#Calculate the coefficients

sqft_model.coefficients
```

```
#Visualising the prediction
#Import matplotlib.pyplot


import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(test_data['sqft_living'],test_data['price'],'.',
         test_data['sqft_living'],sqft_model.predict
         (test_data),'-')

```

```
#Explore other features in data


my_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']
sales[my_features].show()
```

```
turicreate.show(sales['zipcode'],sales['price'])
```

```
#Build a Regression model with more features


my_features_model = turicreate.linear_regression.create(train_data,target="price",features = my_features)
```

```
print(my_features)
```

```
#Comparing two models i.e. sqft_model and my features_model


print(sqft_model.evaluate(test_data))
print(my_features_model.evaluate(test_data))
```
