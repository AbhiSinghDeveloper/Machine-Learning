### ***2. Using OpenCV Library***

## Import ***Libraries***

```
import cv2
```

Load the image using cv2 **Module**

```
img_1 = cv2.imread('/content/drive/My Drive/Colab Notebooks/RGB.jpg')
```

**Check the type of image**

```
type(img_1)
```

**Get the image shape**

```
img_1.shape
```

Let's see the image using **imshow**

```
plt.imshow(img_1)
```

Untill now, we wew working with ***Matplotlib*** and **RGB**

***OpenCV*** is reading the channel as BGR.

We will convert OpenCV to the channels of the photo.

```
#Converting OpenCV, from BGR to RGB

img_1_fix = cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB)
```

```
plt.imshow(img_1_fix)
```

Scale it to Gray and check the Shape

```
#Again load the picture using "imread" and then convert it to Gray scale

img_gray = cv2.imread('/content/drive/My Drive/Colab Notebooks/RGB.jpg')

img_gray.shape
```

```
plt.imshow(img_gray, cmap = 'gray')
```

# **Resize the image**

To resize the image we change the Height Width --> Widht Height

```
img_new = cv2.resize(img_1_fix, (1000, 400))
```

```
plt.imshow(img_new)
```

**Resize with Ratio**

```
width_ratio = 0.5
height_ratio = 0.5
```

```
img_2 = cv2.resize(img_1_fix,(0,0),img_1_fix, width_ratio, height_ratio)
```

```
plt.imshow(img_2)
```

```
#Let's check the shape
#Shape must be half of that of Original.

img_2.shape
```

**FLip on Horizontal Axis**

```
img_3 = cv2.flip(img_1_fix,0)

plt.imshow(img_3)
```

**Flip on Vertical Axis**

```
img_3 = cv2.flip(img_1_fix, 1)

plt.imshow(img_3)
```

***Flip on Horizontal and Vertical Axis***

```
img_3 = cv2.flip(img_1_fix, -1)

plt.imshow(img_3)
```

### ***3. Drawing Shapes on Images***





**Create a Blackimage to work**

```
black_img = np.zeros(shape=(512,513,3), dtype = np.int16)
```

**Get the Shape of the Image**

```
black_img.shape
```

**Show it**

```
plt.imshow(black_img)
```

**Draw a Circle**

```
#The center is the first number on X-axis and second on Y-axis

cv2.circle(img = black_img,
           center = (400,100),
           radius = 50,
           color = (255,0,0),
           thickness = 8)

plt.imshow(black_img)
```

**Filled Circle**

```
cv2.circle(img = black_img,
           center = (400,200),
           radius = 50,
           color = (0,255,0),
           thickness = -1)

plt.imshow(black_img)
```

# ***Draw a Rectangle***

The first number is on X-axis and the second number is on Y-axis.

We need two points, One for 'UP' and one for diagonally 'DOWN'

```
cv2.rectangle(black_img,
              pt1 = (200,200),
              pt2 = (300,300),
              color = (0,255,0),
              thickness = 5)

plt.imshow(black_img)
```

# **Draw Triangle**

```
vertices = np.array([[10,450],
                    [110,350],
                    [180,450]],
                    np.int32)
pts = vertices.reshape(-1,1,2)

cv2.polylines(black_img,
              [pts],
              isClosed=True,
              color = (0,0,255),
              thickness=3)
plt.imshow(black_img)
```
